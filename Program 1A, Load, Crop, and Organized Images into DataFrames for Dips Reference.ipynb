```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import glob, os, shutil
from IPython.core.display import HTML
```

```python
img_list, img_neg = [], []
file_names, file_names_neg = [], []

os.chdir('C:/Users/robin/Programs/Candy_Collect/Grant/Database/Dip_Data/')
for img in glob.glob("*dips.*"):
    img_list.append(cv2.imread(img))
    file_names.append(img[:-4])
    
for img in glob.glob("*neg*.*"):
    img_neg.append(cv2.imread(img))
    file_names_neg.append(img[:-4])
    
print(file_names), print(file_names_neg)
```

```python
folder = 'LossLessFormat'
for filename in os.listdir(folder):
    file_path = os.path.join(folder, filename)
    try:
        if os.path.isfile(file_path) or os.path.islink(file_path):
            os.unlink(file_path)
        elif os.path.isdir(file_path):
            shutil.rmtree(file_path)
    except Exception as e:
        print('Failed to delete %s. Reason: %s' % (file_path, e))
        
# files = glob.glob('Output/*')
# for f in files:
#     os.remove(f)

for i in range(len(img_list)):
    cv2.imwrite('LossLessFormat/' + file_names[i]+'.tiff', img_list[i])
    
for i in range(len(img_neg)):
    cv2.imwrite('LossLessFormat/' + file_names_neg[i]+'.tiff', img_neg[i])
```

```python
df_dips_gray = pd.DataFrame()
```

```python
gray_11_15, gray_11_1, gray_11_22, gray_8_3, gray_9_6 = [], [], [], [], []

for i in range(len(df_dips.dips_11_15)):

    gray_11_15.append((255 - np.average(df_dips.dips_11_15[i], axis = -1))/255)

    gray_11_1.append((255 - np.average(df_dips.dips_11_1[i], axis = -1))/255)

    gray_11_22.append((255 - np.average(df_dips.dips_11_22[i], axis = -1))/255)

    gray_8_3.append((255 - np.average(df_dips.dips_8_3[i], axis = -1))/255)
    
    gray_9_6.append((255 - np.average(df_dips.dips_9_6[i], axis = -1))/255)
```

```python
df_dips_gray['dips_11_15'] = gray_11_15
df_dips_gray['dips_11_1'] = gray_11_1
df_dips_gray['dips_11_22'] = gray_11_22
df_dips_gray['dips_8_3'] = gray_8_3
df_dips_gray['dips_9_6'] = gray_9_6
```

```python
fig, axs = plt.subplots(2, 5, figsize=(10, 2))

for i in range(len(df_dips_gray.dips_11_15)):
    axs[i, 0].imshow(df_dips_gray.iloc[[i]].dips_11_15[i], cmap = 'Greys_r')

for i in range(len(df_dips_gray.dips_11_1)):
    axs[i, 1].imshow(df_dips_gray.iloc[[i]].dips_11_1[i], cmap = 'Greys_r')
    
for i in range(len(df_dips_gray.dips_11_22)):
    axs[i, 2].imshow(df_dips_gray.iloc[[i]].dips_11_22[i], cmap = 'Greys_r')
    
for i in range(len(df_dips_gray.dips_8_3)):
    axs[i, 3].imshow(df_dips_gray.iloc[[i]].dips_8_3[i], cmap = 'Greys_r')
    
for i in range(len(df_dips_gray.dips_9_6)):
    axs[i, 4].imshow(df_dips_gray.iloc[[i]].dips_9_6[i], cmap = 'Greys_r')
    
plt.tight_layout()
```
