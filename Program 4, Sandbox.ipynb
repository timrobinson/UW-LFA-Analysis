```python
df['avg'] = df[['N1', 'N2', 'N3', 'N4']].mean(axis=1)
df['stnd'] = df[['N1', 'N2', 'N3', 'N4']].std(axis=1)
df['err'] = df['stnd']/4
df['threshold'] = df.avg[0] + 3*df.stnd[0]
df['bacterial_load'] = [0, 1*10**5, 1.5*10**5, 5*10**5, 1*10**6, 5*10**6, 1*10**7, 1*10**9]
```

```python
fig = plt.figure()

def set_axis_style(ax, labels):
    ax.set_xticks(np.arange(1, len(labels) + 1), labels=labels)
    ax.set_xlim(0.25, len(labels) + 1)
    ax.set_xlabel('\nBacterial Load')

ax = fig.add_axes([0,0,1,1])

ax.set_xscale("log")

set_axis_style(ax, labels)
    
bp = ax.violinplot(data_to_plot, showmeans = True, showextrema=False)
plt.grid()

plt.show()
```
